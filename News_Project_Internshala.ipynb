```python
from sklearn.model_selection import train_test_split # used to split the given data into training and testing 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer # used for vectoriation TF-tern Frequency, IDF-inverse document frequency
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
PassiveAggressiveClassifier
```

```python
x_train,x_test,y_train,y_test= train_test_split(news["text"], y, test_size=0.3, random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
tfidf_vectorizer=TfidfVectorizer(stop_words='english', max_df=0.7)
```

```python
tfidf_vectorizer
```

```python
tfidf_train=tfidf_vectorizer.fit_transform(x_train)
tfidf_test=tfidf_vectorizer.transform(x_test)
```

```python
tfidf_train
```

```python
tfidf_test
```

```python
pca=PassiveAggressiveClassifier(max_iter= 50)
```

```python
pca
```

```python
pca.fit(tfidf_train,y_train)
```

```python
y_pred=pca.predict(tfidf_test)
```

```python
y_pred[3]
```

```python
y[3]
```

```python
score=accuracy_score(y_test,y_pred)
```

```python
score
```

```python
confusion_matrix(y_test,y_pred, labels=['FAKE','REAL'])
```
